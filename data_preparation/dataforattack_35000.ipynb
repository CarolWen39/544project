```python
data = pd.read_csv("/content/drive/MyDrive/544_project/data/raw_data/3w5.csv", index_col=0)
```

```python
onew5 = pd.read_csv("/content/drive/MyDrive/544_project/data/raw_data/1w5.csv",index_col=0)
```

```python
data
```

```python
newdata = data[['rating_alternateName_normalized', 'claimReview_claimReviewed']]
newdata
```

```python
MIXTURE = newdata[newdata['rating_alternateName_normalized']=='MIXTURE']
```

```python
OTHER = newdata[newdata['rating_alternateName_normalized']=='OTHER']
```

```python
FALSE = newdata[newdata['rating_alternateName_normalized']=='FALSE']
```

```python
TRUE = newdata[newdata['rating_alternateName_normalized']=='TRUE']
TRUE
```

```python
out1 = pd.concat([MIXTURE,OTHER,FALSE,TRUE])
out1.columns = ['label','text']
out1 = out1.reset_index()
out1.drop(columns='index')
```

```python
onew5 = onew5.reset_index()
onew5.drop(columns='index')
```

```python
output = pd.concat([out1,onew5])
output
```

```python
output = output.sample(frac=1).reset_index(drop=True)
output.drop(columns="index")
```

```python
output.loc[output['label'] == "MIXTURE", 'label'] = 'Mixture'
output.loc[output['label'] == "OTHER", 'label'] = 'Other'
output.loc[output['label'] == "TRUE", 'label'] = 'True'
output.loc[output['label'] == "FALSE", 'label'] = 'False'
```

```python
output.to_csv("/content/drive/MyDrive/544_project/data/raw_data/finaldata.csv")
```
