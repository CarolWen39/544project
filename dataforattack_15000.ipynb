```python
import pandas as pd
import numpy as np
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from google.colab import drive
drive.mount('/content/drive')
```

```python
data_6k = pd.read_csv("/content/drive/MyDrive/544_project/data/raw_data/fake_or_real_news.csv")
data_6k = data_6k[['text','label']]
data_6k[data_6k['label']=='FAKE'] = "FALSE"
data_6k[data_6k['label']=='REAL'] = "TRUE"
data_6k
```

```python
data_true = pd.read_csv("/content/drive/MyDrive/544_project/data/raw_data/True.csv")
data_true = data_true.drop(['title','subject', 'date'], axis=1)
data_true['label'] = 'TRUE'
data_true
```

```python
data_false = pd.read_csv("/content/drive/MyDrive/544_project/data/raw_data/Fake.csv")
data_false = data_false.drop(['title','subject', 'date'], axis=1)
data_false['label'] = 'FALSE'
data_false
```

```python
data = pd.concat([data_6k, data_false, data_true])
data
```

```python
data = data.sample(frac=1, random_state= 44).reset_index(drop=True)
data
```

```python
tmp = data[data['label']=='FALSE']
```

```python
tmp
```

```python
new_f = tmp.sample(n=5000)
```

```python
new_f
```

```python
new_t = data[data['label']=='TRUE']
```

```python
new_t = new_t.sample(10000)
```

```python
new_t
```

```python
df_1w5 = pd.concat([new_f,new_t])
```

```python
df_1w5
```

```python
count = 0
for i in (df_1w5["text"]):
  if count == 2:
    # print(i)
    print(len(i))
    break
  count += 1
```

```python
lst = []
for i in df_1w5["text"]:
  # print(i)
  if i != "text" or i != "FALSE":
    tmp = []
    tmp = i.split(".")
    lst.append(tmp[0])
```

```python
output_df = pd.DataFrame(lst,columns = ['text'])
```
